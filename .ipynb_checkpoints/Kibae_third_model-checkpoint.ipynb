```python
# Your code here - remember to use markdown cells for comments as well!
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
```

```python
df = pd.read_csv('data/kc_house_data.csv')
```

```python
payload=pd.read_html("https://www.zipcodestogo.com/Washington/")
df_zipcode = payload[1]
df_zipcode = df_zipcode.drop([0,1], axis=0)
df_zipcode = df_zipcode.drop([3], axis=1)
df_zipcode.columns = ['zipcode', 'city', 'county']
df_zipcode.index -= 2

zipcode_dic = {}
for i in range(len(df_zipcode)):
    zipcode_dic[int(df_zipcode.zipcode[i])] = df_zipcode.city[i]
```

```python
df['city'] = df['zipcode'].map(zipcode_dic)
```

```python
df.drop(['id', 'view','sqft_above', 'sqft_basement', 'lat', 'long', 'sqft_living15', 'sqft_lot15'], axis=1, inplace=True)
df.dropna(subset=['yr_renovated', 'waterfront'], inplace=True)
```

```python
df['year'] = pd.DatetimeIndex(df['date']).year
df['month'] = pd.DatetimeIndex(df['date']).month
```

```python
df['renovated_period'] = df.yr_renovated - df.yr_built
```

```python
df_renovated = df[df['renovated_period'] > 0]
```

```python
df['yr_from_renovated'] = df.year - df.yr_renovated
```

```python
def f_0_20(row):
    if row['yr_renovated'] != 0:
        if row['year'] - row['yr_renovated'] <= 20:
            val = 1
        else:
            val = 0
    else:
        val = 0
    return val


def f_20_over(row):
    if row['yr_renovated'] != 0:
        if row['year'] - row['yr_renovated'] > 20:
            val = 1
        else:
            val = 0
    else:
        val = 0
    return val



df['renov_0_20yr'] = df.apply(f_0_20, axis=1)
df['renov_20yr_over'] = df.apply(f_20_over, axis=1)
```

```python
high_price_cities = ['Sammamish', 'Bellevue', 'Medina', 'Mercer Island']
low_price_cities = ['Auburn', 'Enumclaw', 'Federal Way', 'Kent', 'Maple Valley']
```

```python
def high_price_city(row):
    if row['city'] in high_price_cities:
        value = 1
    else:
        value = 0
    return value

def low_price_city(row):
    if row['city'] in low_price_cities:
        value = 1
    else:
        value = 0
    return value

df['high_price_cities'] = df.apply(high_price_city, axis=1)
df['low_price_cities'] = df.apply(low_price_city, axis=1)
```

```python
lr_model = ols(formula='price ~ bedrooms + bathrooms + sqft_living + waterfront + condition + grade + yr_built +renov_0_20yr + renov_20yr_over + high_price_cities + low_price_cities', data=df).fit()
lr_model.summary()
```

```python
df['predicted_price'] = (5.371e+06) - (4.255e+04)*df.bedrooms + (5.749e+04)*df.bathrooms + (178.1664)*df.sqft_living + (7.581e+05)*df.waterfront + (1.969e+04)*df.condition + (1.065e+05)*df.grade - (3072.8645)*df.yr_built + (6.364e+04)*df.renov_0_20yr - (3.011e+04)*df.renov_20yr_over + (1.352e+05)*df.high_price_cities - (1.594e+05)*df.low_price_cities
```

```python
df['bedrooms_squared'] = df.bedrooms*df.bedrooms
df['bathrooms_squared'] = df.bathrooms*df.bathrooms
df['grade_squared'] = df.grade*df.grade
```

```python
lr_model = ols(formula='price ~ bedrooms + bathrooms_squared + sqft_living + waterfront + condition + grade_squared + yr_built +renov_0_20yr + renov_20yr_over + high_price_cities + low_price_cities', data=df).fit()
lr_model.summary()
```

```python
df['total_sqft'] = df['sqft_living'] + df['sqft_lot']
```

```python
lr_model = ols(formula='price ~ bedrooms + bathrooms_squared + total_sqft + waterfront + condition + grade_squared + yr_built +renov_0_20yr + renov_20yr_over + high_price_cities + low_price_cities', data=df).fit()
lr_model.summary()
```
